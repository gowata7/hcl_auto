```python
from datetime import datetime, timedelta
from calendar import monthrange

def period(year, month):
    first_day = datetime(year, month, 1)
    last_day = first_day + timedelta(days=calendar.monthrange(year, month)[1])
    return first_day.strftime('%Y-%m-%d'), last_day.strftime('%Y-%m-%d %H:%M:%S')

first_date, last_date = period(int(year), int(month))
print(first_date, last_date)

next_month = int(month) + 1
first_date_next, last_date_next = period(int(year), next_month)
print(first_date_next, last_date_next)
```

```python
import httpx

async def login(username, password):
    async with httpx.AsyncClient() as client:
        login_data = {"username": username, "password": password}
        login_response = await client.post("http://10.11.66.85:8080/users/login", data=login_data)
        
        if login_response.status_code == 200:
            access_token = login_response.json().get("access_token")
            return access_token
        else:
            return None

async def fetch_data(access_token, data_endpoint, request_header):
    async with httpx.AsyncClient() as client:
        headers = {
            "Authorization": f"Bearer {access_token}",
            "Content-Type": "application/json",
        }
        data_response = await client.post(data_endpoint, headers=headers, json=request_header)
        
        if data_response.status_code == 200:
            fetched_data = data_response.json()
            return fetched_data
        else:
            return None

async def login_and_fetch_data(username, password, data_endpoint):
    access_token = await login(username, password)
    
    if access_token:
        fetched_data = await fetch_data(access_token, data_endpoint)
        return fetched_data
    else:
        return "Login failed"

# Replace 'username_here' and 'password_here' with actual credentials
username = 'test2@test2.com'
password = '2222'

access_token = await login(username, password)
```
